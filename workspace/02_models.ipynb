# OpenUBA SDK: Model Management

This notebook demonstrates model registration, versioning, and the model hub.

**SDK functions covered:** `register_model`, `publish_version`, `load_model`, `list_models`, `get_model`, `install`, `uninstall`, `list_installed`, `run`

```python
import openuba
print(f'OpenUBA SDK v{openuba.__version__}')
```

## 1. Register a Model

Register a model with the platform via the SDK API.

```python
# Register a model by name and framework
try:
    model = openuba.register_model(
        name='nb-anomaly-detector',
        framework='sklearn',
        description='Isolation Forest anomaly detector registered from notebook',
    )
    model_id = model.get('id') or model.get('model_id')
    print(f'Model registered: {model}')
    print(f'Model ID: {model_id}')
except Exception as e:
    print(f'register_model: {e}')
    model_id = None
```

## 2. Register a Model with Auto-Detection

Pass a live sklearn model object for automatic framework detection and serialization.

```python
from sklearn.ensemble import IsolationForest

# Create a real sklearn model
clf = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
print(f'Created IsolationForest: {clf}')

# Register with auto-detection (detects sklearn, serializes model)
try:
    auto_model = openuba.register_model(
        name='nb-auto-iforest',
        model=clf,
        description='Auto-detected sklearn IsolationForest',
    )
    auto_model_id = auto_model.get('id') or auto_model.get('model_id')
    print(f'Auto-registered model: {auto_model}')
except Exception as e:
    print(f'register_model (auto): {e}')
    auto_model_id = None
```

## 3. Publish a Model Version

```python
if model_id:
    try:
        version = openuba.publish_version(
            model_id=model_id,
            version='1.0.0',
            summary='Initial release of anomaly detector',
        )
        print(f'Version published: {version}')
    except Exception as e:
        print(f'publish_version: {e}')
else:
    print('Skipped: no model_id from registration')
```

## 4. Load a Model by Name

```python
try:
    loaded = openuba.load_model('nb-anomaly-detector')
    print(f'Loaded model: {loaded}')
except Exception as e:
    print(f'load_model: {e}')
```

## 5. Model Hub: List Registry Models

The model hub fetches from the openuba-model-hub GitHub registry.

```python
try:
    hub_models = openuba.list_models()
    print(f'Hub models available: {len(hub_models)}')
    for m in hub_models[:5]:
        print(f'  - {m.get("name")}: {(m.get("description") or "")[:60]}')
except Exception as e:
    print(f'list_models (hub): {e}')
    print('(Hub requires internet access to openuba.org registry)')
```

## 6. Model Hub: Get Model Details

```python
try:
    model_info = openuba.get_model('ssh-anomaly-detector')
    print(f'Model info: {model_info}')
except Exception as e:
    print(f'get_model: {e}')
```

## 7. List Installed Models

```python
try:
    installed = openuba.list_installed()
    print(f'Locally installed models: {len(installed)}')
    for m in installed:
        print(f'  - {m["name"]} at {m["path"]}')
except Exception as e:
    print(f'list_installed: {e}')
```
