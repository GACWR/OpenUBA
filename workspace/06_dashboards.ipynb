# OpenUBA SDK: Dashboards

This notebook demonstrates creating, updating, and listing dashboards.

**SDK functions covered:** `create_dashboard`, `update_dashboard`, `list_dashboards`

```python
import openuba
print(f'OpenUBA SDK v{openuba.__version__}')
```

## 1. Create a Dashboard with Layout

```python
dashboard_id = None
try:
    dashboard = openuba.create_dashboard(
        name='nb-security-overview',
        description='Security monitoring dashboard created from notebook',
        layout=[
            {
                'title': 'Anomaly Distribution',
                'chart_type': 'bar',
                'x_key': 'hour',
                'y_key': 'count',
                'data': [
                    {'hour': '00:00', 'count': 12},
                    {'hour': '06:00', 'count': 3},
                    {'hour': '12:00', 'count': 8},
                    {'hour': '18:00', 'count': 25},
                ],
            },
            {
                'title': 'Risk by Entity Type',
                'chart_type': 'pie',
                'x_key': 'type',
                'y_key': 'count',
                'data': [
                    {'type': 'User', 'count': 45},
                    {'type': 'Service', 'count': 20},
                    {'type': 'Device', 'count': 15},
                ],
            },
        ],
    )
    dashboard_id = dashboard.get('id')
    print(f'Dashboard created:')
    for k, v in dashboard.items():
        print(f'  {k}: {v}')
except Exception as e:
    print(f'create_dashboard: {e}')
```

## 2. Create a Minimal Dashboard

```python
try:
    minimal = openuba.create_dashboard(
        name='nb-empty-dashboard',
        description='Empty dashboard for later customization',
    )
    print(f'Minimal dashboard: {minimal}')
except Exception as e:
    print(f'create_dashboard (minimal): {e}')
```

## 3. Update a Dashboard

```python
if dashboard_id:
    try:
        updated = openuba.update_dashboard(
            dashboard_id=dashboard_id,
            name='nb-security-overview-v2',
            layout=[
                {
                    'title': 'Updated Anomaly Distribution',
                    'chart_type': 'line',
                    'x_key': 'hour',
                    'y_key': 'count',
                    'data': [
                        {'hour': '00:00', 'count': 15},
                        {'hour': '04:00', 'count': 5},
                        {'hour': '08:00', 'count': 22},
                        {'hour': '12:00', 'count': 18},
                        {'hour': '16:00', 'count': 30},
                        {'hour': '20:00', 'count': 42},
                    ],
                },
            ],
        )
        print(f'Dashboard updated: {updated}')
    except Exception as e:
        print(f'update_dashboard: {e}')
else:
    print('Skipped: no dashboard_id')
```

## 4. List All Dashboards

```python
try:
    all_dashboards = openuba.list_dashboards()
    print(f'Total dashboards: {len(all_dashboards)}')
    for d in all_dashboards:
        panels = len(d.get('layout', []))
        print(f'  - {d.get("name")}: {panels} panel(s)')
except Exception as e:
    print(f'list_dashboards: {e}')
```
