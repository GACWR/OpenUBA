# OpenUBA SDK: UBA Security Queries & Data Connectors

This notebook demonstrates querying anomalies, entities, cases, rules, and data connectors.

**SDK functions covered:** `query_anomalies`, `get_entity_risk`, `query_cases`, `list_rules`, `query_spark`, `query_elasticsearch`

```python
import openuba
print(f'OpenUBA SDK v{openuba.__version__}')
```

## 1. Query Anomalies

Search for detected anomalies with optional filters.

```python
# Query all anomalies
try:
    anomalies = openuba.query_anomalies(limit=10)
    if isinstance(anomalies, list):
        print(f'Anomalies found: {len(anomalies)}')
        for a in anomalies[:5]:
            print(f'  - entity={a.get("entity_id")}, risk={a.get("risk_score")}')
    else:
        print(f'Anomalies response: {anomalies}')
except Exception as e:
    print(f'query_anomalies: {e}')
```

```python
# Query high-risk anomalies only
try:
    high_risk = openuba.query_anomalies(min_risk=0.8, limit=5)
    if isinstance(high_risk, list):
        print(f'High-risk anomalies (risk >= 0.8): {len(high_risk)}')
    else:
        print(f'High-risk response: {high_risk}')
except Exception as e:
    print(f'query_anomalies (high-risk): {e}')
```

```python
# Query anomalies for a specific entity
try:
    entity_anomalies = openuba.query_anomalies(entity_id='user-001', limit=5)
    if isinstance(entity_anomalies, list):
        print(f'Anomalies for user-001: {len(entity_anomalies)}')
    else:
        print(f'Entity anomalies response: {entity_anomalies}')
except Exception as e:
    print(f'query_anomalies (entity): {e}')
```

## 2. Get Entity Risk Profile

```python
try:
    risk = openuba.get_entity_risk('user-001')
    print(f'Entity risk profile:')
    if isinstance(risk, dict):
        for k, v in risk.items():
            print(f'  {k}: {v}')
    else:
        print(f'  {risk}')
except Exception as e:
    print(f'get_entity_risk: {e}')
```

## 3. Query Security Cases

```python
# Query all open cases
try:
    cases = openuba.query_cases(status='open', limit=10)
    if isinstance(cases, list):
        print(f'Open cases: {len(cases)}')
        for c in cases[:5]:
            print(f'  - {c.get("title", "untitled")}: severity={c.get("severity")}')
    else:
        print(f'Cases response: {cases}')
except Exception as e:
    print(f'query_cases: {e}')
```

```python
# Query critical cases
try:
    critical = openuba.query_cases(severity='critical', limit=5)
    if isinstance(critical, list):
        print(f'Critical cases: {len(critical)}')
    else:
        print(f'Critical cases response: {critical}')
except Exception as e:
    print(f'query_cases (critical): {e}')
```

## 4. List Detection Rules

```python
try:
    rules = openuba.list_rules(enabled=True)
    if isinstance(rules, list):
        print(f'Enabled detection rules: {len(rules)}')
        for r in rules[:5]:
            print(f'  - {r.get("name", "unnamed")}: {r.get("description", "")}')
    else:
        print(f'Rules response: {rules}')
except Exception as e:
    print(f'list_rules: {e}')
```

## 5. Spark SQL Query

Execute a Spark SQL query using PySpark (local mode).

```python
try:
    # Create a simple Spark SQL query with inline data
    results = openuba.query_spark("SELECT 1 as id, 'test' as name")
    print(f'Spark SQL results: {results}')
except ImportError as e:
    print(f'query_spark (import): {e}')
except Exception as e:
    print(f'query_spark: {e}')
```

## 6. Elasticsearch Query

Query Elasticsearch (demonstrates API even if ES is not available).

```python
try:
    results = openuba.query_elasticsearch(
        index='auth-logs-*',
        query_body={'query': {'match_all': {}}, 'size': 5},
    )
    print(f'Elasticsearch results: {len(results)} hits')
except ImportError as e:
    print(f'query_elasticsearch (import): {e}')
except Exception as e:
    print(f'query_elasticsearch (connection): {e}')
    print('(Expected: Elasticsearch not available in this environment)')
```
