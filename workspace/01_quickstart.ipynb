# OpenUBA SDK Quickstart

This notebook demonstrates the core setup and configuration of the OpenUBA Python SDK.

**SDK functions covered:** `configure`, `OpenUBAClient`, `__version__`

```python
import openuba
print(f'OpenUBA SDK v{openuba.__version__} loaded successfully!')
```

## Configure the SDK

The SDK auto-configures from environment variables (`OPENUBA_API_URL`, `OPENUBA_TOKEN`).
You can also configure explicitly:

```python
import os

# Show current environment configuration
api_url = os.environ.get('OPENUBA_API_URL', 'http://localhost:8000')
workspace_id = os.environ.get('OPENUBA_WORKSPACE_ID', 'not set')
print(f'API URL: {api_url}')
print(f'Workspace ID: {workspace_id}')

# Configure the SDK (uses env vars automatically)
openuba.configure(api_url=api_url)
print('\nSDK configured successfully!')
```

## OpenUBAClient Class

The `OpenUBAClient` class is the core of the SDK. Module-level functions are convenience wrappers.

```python
# Direct client instantiation
client = openuba.OpenUBAClient(api_url=api_url)
print(f'Client class: {type(client).__name__}')
print(f'API URL: {client.api_url}')
print(f'Model dir: {client.model_dir}')
```

## SDK Exports

All public functions and classes available in the SDK:

```python
print(f'Total exported items: {len(openuba.__all__)}')
print()
for item in openuba.__all__:
    obj = getattr(openuba, item)
    kind = 'class' if isinstance(obj, type) else 'function'
    print(f'  openuba.{item} ({kind})')
```

## Context Classes

The SDK provides execution contexts for model training and visualization rendering:

```python
# ModelContext - used during training/inference in runner pods
ctx = openuba.ModelContext(job_id='demo-job', model_id='demo-model', run_type='train')
print(f'ModelContext created:')
print(f'  job_id: {ctx.job_id}')
print(f'  model_id: {ctx.model_id}')
print(f'  run_type: {ctx.run_type}')

# Log a metric (buffered, will attempt API flush)
ctx.log_metric('accuracy', 0.95, epoch=1)
ctx.log('Training started', level='info')
print('  Logged metric and message successfully')

# VisualizationContext - used for rendering figures
print(f'\nVisualizationContext backends: {list(openuba.VisualizationContext.BACKEND_OUTPUT_MAP.keys())}')
```
