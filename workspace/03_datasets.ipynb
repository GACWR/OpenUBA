# OpenUBA SDK: Dataset Management

This notebook demonstrates creating, listing, and retrieving datasets.

**SDK functions covered:** `create_dataset`, `list_datasets`, `get_dataset`

```python
import openuba
print(f'OpenUBA SDK v{openuba.__version__}')
```

## 1. Create Datasets

Create dataset records for different data sources and formats.

```python
datasets_created = []

# Create a CSV dataset
try:
    ds1 = openuba.create_dataset(
        name='nb-auth-logs-csv',
        description='Authentication logs in CSV format',
        source_type='upload',
        format='csv',
    )
    datasets_created.append(ds1)
    print(f'CSV dataset created: {ds1}')
except Exception as e:
    print(f'create_dataset (csv): {e}')
```

```python
# Create a Parquet dataset
try:
    ds2 = openuba.create_dataset(
        name='nb-network-flows-parquet',
        description='Network flow data in Parquet format',
        source_type='s3',
        format='parquet',
    )
    datasets_created.append(ds2)
    print(f'Parquet dataset created: {ds2}')
except Exception as e:
    print(f'create_dataset (parquet): {e}')
```

```python
# Create a JSON dataset
try:
    ds3 = openuba.create_dataset(
        name='nb-syslog-json',
        description='System logs in JSON format',
        source_type='upload',
        format='json',
    )
    datasets_created.append(ds3)
    print(f'JSON dataset created: {ds3}')
except Exception as e:
    print(f'create_dataset (json): {e}')
```

## 2. List All Datasets

```python
try:
    all_datasets = openuba.list_datasets()
    print(f'Total datasets: {len(all_datasets)}')
    for ds in all_datasets:
        print(f'  - {ds.get("name")}: format={ds.get("format")}, source={ds.get("source_type")}')
except Exception as e:
    print(f'list_datasets: {e}')
```

## 3. Get Dataset by ID

```python
if datasets_created:
    ds_id = datasets_created[0].get('id')
    try:
        ds_detail = openuba.get_dataset(ds_id)
        print(f'Dataset detail:')
        for k, v in ds_detail.items():
            print(f'  {k}: {v}')
    except Exception as e:
        print(f'get_dataset: {e}')
else:
    print('Skipped: no datasets were created')
```

```python
print('\n=== DATASET MANAGEMENT COMPLETE ===')
print(f'Datasets created: {len(datasets_created)}')
print('Functions tested: create_dataset, list_datasets, get_dataset')
print('All dataset management checks passed!')
```
