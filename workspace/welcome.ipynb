# Welcome to OpenUBA Workspace

This is your interactive environment for developing User Behavior Analytics models,
creating visualizations, and building dashboards.

## Getting Started

The `openuba` Python package is pre-installed and auto-configured in this workspace.

````python
import openuba

# The SDK is auto-configured via environment variables
# OPENUBA_API_URL and OPENUBA_TOKEN are set automatically
print(f"OpenUBA SDK ready")
````

## Available Features

### Model Management
```python
# Register a model
from sklearn.ensemble import IsolationForest
clf = IsolationForest(n_estimators=100)
handle = openuba.register_model('ssh-anomaly-detector', model=clf)

# Train remotely
job = openuba.start_training(handle['model_id'], dataset_id='ssh-logs')
```

### Visualizations
```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(data, bins=50)
viz = openuba.create_visualization('risk-dist', backend='matplotlib')
```

### Data Access
```python
# Query anomalies
anomalies = openuba.query_anomalies(min_risk=0.7)

# Query Elasticsearch
df = openuba.query_elasticsearch('ssh-*', {'query': {'match_all': {}}})
```
